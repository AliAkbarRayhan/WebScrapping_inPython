```python
import requests
import pandas as pd
import json
```

```python
data = []
for page in range(1,11):
    url= f'https://www.daraz.com.bd/smartphones/?ajax=true&page={page}&spm=a2a0e.home.cate_6.1.573712f7b48VtW'
    response =  requests.get(url)
    data_json = response.json()
    print(f'{page} no page is done')
    all_phones = data_json['mods']['listItems']
    for item,phone in enumerate(all_phones):
        try:
            originalPrice = phone['originalPrice']
        except:
            originalPrice = phone['price']

        try:
            discount = phone['discount']
        except:
            discount = 'Not available'
        try:
            rating = phone['ratingScore']
        except:
            rating = 'Not available'
        d = {
            'name':phone['name'],
            'originalPrice': originalPrice,
            'price':phone['price'],
            'discount':discount,
            'ratingScore': rating,
            'review':phone['review'],
            'brandName':phone['brandName']

        }
        data.append(d)
    
```
